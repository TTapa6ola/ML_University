```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('eustock.csv')
y = np.arange(len(data['DAX']))
```

```python
variants = [
    'DAX',
    'SMI',
    'CAC',
    'FTSE',
]
scores = []

for var in variants:
    train_X, test_X, train_y, test_y = train_test_split(data[var].to_numpy().reshape(-1, 1), y, train_size=0.8, random_state=42)
    reg = LinearRegression().fit(train_X, train_y)
    scores.append(reg.score(test_X, test_y))
    plt.plot(y, data[var], label=var)
    print(var, reg.coef_)

train_X, test_X, train_y, test_y = train_test_split(data, y, train_size=0.8, random_state=42)
reg = LinearRegression().fit(train_X, train_y)
scores.append(reg.score(test_X, test_y))
print('All', reg.coef_)

plt.legend(loc='best')
plt.show()
variants.append('All')
```

```python
plt.bar(variants, scores)
plt.xlabel('Stock')
plt.ylabel('Score')
plt.grid()
plt.show()
```
