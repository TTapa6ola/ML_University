```python
import torch
from torch.autograd import Variable
from torch import nn, optim
import torch.nn.functional as F
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data0 = pd.read_csv('nn_0.csv')
X0_df = data0.drop('class', axis=1)
y0_df = data0['class']

X0 = torch.tensor(X0_df.values)
y0 = torch.tensor(y0_df.values)


X0_train, X0_test, y0_train, y0_test = train_test_split(X0, y0, train_size=0.8)
```

```python
plt.scatter(X0[:, 0], X0[:, 1], c=y0, cmap=plt.cm.rainbow)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('nn_0.csv')
plt.show()
```

```python
epoch = 100

nn_relu = nn.Sequential(
    nn.Linear(2, 1),
    nn.ReLU()
)

optim_SGD = optim.SGD(nn_relu.parameters(), lr=3e-3)
loss_f = nn.MSELoss()
losses = []

for i in range(epoch):
    pred = nn_relu(X0_train.float())
    loss = loss_f(pred, y0_train.float())
    losses.append(loss)
    
    optim_SGD.zero_grad()
    loss.backward()
    optim_SGD.step()
```
